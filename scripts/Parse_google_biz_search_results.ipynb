```python
import pandas as pd
import re
```

```python
# FIPS codes
file_path = '/Users/rsandidge/Dropbox/jupyter notebook files/Python_Projects/Market_Analysis_Rafts/copypaste_text_surfshops'
with open(file_path, 'r') as file:
    text = file.read()  # Reads the entire file
    print(text)
```

```python
# Split into sections for each shop
shops = [shop.strip() for shop in text.split("DIRECTIONS") if shop.strip()] #splits text into substrings at "DIRECTIONS" and strips leading and trailing spaces
```

```python
print(shops)
```

```python
# Prepare lists for columns
shop_names = []
shop_types = []
cities = []
phone_numbers = []

# Parse each shop section
for shop in shops:
    lines = shop.splitlines()
    
    # Extract required information using regex and position
    shop_name = lines[0].strip() if len(lines) > 0 else ""
    shop_type = re.search(r'· (.+)', lines[1]).group(1).strip() if len(lines) > 1 and '·' in lines[1] else ""
    city_state = lines[2].strip() if len(lines) > 2 else ""
    phone = re.search(r'\((\d{3}\) \d{3}-\d{4})', shop).group(1) if re.search(r'\(\d{3}\) \d{3}-\d{4}', shop) else ""
    
    # Append to lists
    shop_names.append(shop_name)
    shop_types.append(shop_type)
    cities.append(city_state)
    phone_numbers.append(phone)

# Create a DataFrame
df = pd.DataFrame({
    "Shop Name": shop_names,
    "Shop Type": shop_types,
    "City, State": cities,
    "Phone Number": phone_numbers
})

#split city and state into separate columns

# Display the DataFrame
display(df)
```
